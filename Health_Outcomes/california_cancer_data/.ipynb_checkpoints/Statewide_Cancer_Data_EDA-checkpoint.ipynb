```python
import pandas as pd
import numpy as np
import os

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
first  = pd.read_csv('./statewide_rates/brain_and_nervous.csv')
first.columns
```

```python
cancer_rates = pd.DataFrame(columns=first.columns)
```

```python
for record in os.listdir('./statewide_rates/'):
    file_df = pd.read_csv('./statewide_rates/' + record)
    cancer_rates = pd.concat([cancer_rates, file_df], axis=0)
```

```python
cancer_rates
```

```python
fig, ax = plt.subplots(figsize=(17, 10))
sns.lineplot(data=cancer_rates[cancer_rates['Cancer Type'] != 'All Types of Cancer'], 
             x = 'Year', 
             y = 'Age-Adjusted Rate', 
             hue='Cancer Type')
plt.show()
```
